```python
import requests
BERLIN_LATITUDE=52.5065133
BERLIN_LONGITUDE=13.144549
with open('darksky_api_key') as f: DARKSKY_API_KEY=f.read().strip()
DARKSKY_BASE_URL=f'https://api.darksky.net/forecast/{DARKSKY_API_KEY}/{BERLIN_LATITUDE},{BERLIN_LONGITUDE},'
```

```python
import datetime
def get_api_timestamp_from_date(ymd_string):
    year,month,day=[int(number_string) for number_string in ymd_string.split('-')]
    return datetime.datetime(year,month,day,12,0,0).isoformat()
```

```python
get_api_timestamp_from_date('2011-05-20')
```

```python
def get_weather_data_given_date(ymd_string):
    return requests.get(DARKSKY_BASE_URL+get_api_timestamp_from_date(ymd_string)).json()
```

```python
get_weather_data_given_date("2019-05-19")
```

```python
precip = []
for x in df_dummy['Date']:
    precip.append(get_weather_data_given_date(x)['currently']['icon'])
```

```python
precip
```

```python
rained = []
for x in precip:
    if x == 'rain':
        rained.append("True")
    else:
        rained.append("False")
rained
```

```python
home_win = []
away_win = []
for outcome in df_dummy['FTR']:
    if outcome == 'A':
        home_win.append(0)
        away_win.append(1)
    elif outcome == 'H':
        home_win.append(1)
        away_win.append(0)
    else:
        home_win.append(0)
        away_win.append(0)
        
```

```python
df_dummy['home_win'] = home_win
df_dummy['away_win'] = away_win
```

```python
home_loss = []
away_loss = []
for outcome in df_dummy['FTR']:
    if outcome == 'A':
        home_loss.append(1)
        away_loss.append(0)
    elif outcome == 'H':
        home_loss.append(0)
        away_loss.append(1)
    else:
        home_loss.append(0)
        away_loss.append(0)
```
